##Torneos Internacionales, Clasificados

```python
torneos_internacionales = datos[datos.Pos <= 5]
torneos_internacionales.head(10)
```

```python
torneos_internacionales[torneos_internacionales['Pos'] == 5]
```

```python
promedio_para_clasificar = torneos_internacionales[torneos_internacionales['Pos'] == 5].Pts.mean().round(2)
```

```python
sns.barplot(x = 'Team', y = 'Pts',data = torneos_internacionales )
```

```python
conteo = torneos_internacionales['Team'].value_counts()
```

```python
sns.catplot(x = 'Team', kind = 'count', data = torneos_internacionales, aspect = 3, order = conteo.index, palette = 'rocket')
plt.title('The most qualified to international competitions', fontsize = 20)
plt.ylabel('Total')
```
